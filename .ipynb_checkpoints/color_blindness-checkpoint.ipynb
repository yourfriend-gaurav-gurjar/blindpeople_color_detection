```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("pic.jpg")
```

```python
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,5))
plt.imshow(image_RGB)
```

```python
image_hsv = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2HSV)
```

```python
red_lower_color = np.array([136, 87, 111], np.uint8)
red_upper_color = np.array([180,255,255], np.uint8)
pink_lower_color = np.array([130, 0, 220], np.uint8)
pink_upper_color = np.array([170, 255, 255], np.uint8)
green_lower_color = np.array([38, 138, 40], np.uint8)
green_upper_color = np.array([70, 255, 255], np.uint8)
```

```python
red_mask = cv2.inRange(image_hsv, red_lower_color, red_upper_color)
pink_mask = cv2.inRange(image_hsv, pink_lower_color, pink_upper_color)
green_mask = cv2.inRange(image_hsv, green_lower_color, green_upper_color)
```

```python
#start from here
mask = red_mask + pink_mask + green_mask
plt.figure(figsize=(20,5))
plt.imshow(mask)
```

```python
res = cv2.bitwise_and(image_RGB, image_RGB, mask=mask)
plt.figure(figsize=(20,5))
plt.imshow(res)
```

```python
# New Lines
light_brown_lower_color = np.array([134,119,14], np.uint8) # from Red to LightBrown
light_brown_upper_color = np.array([154,139,94], np.uint8) # from Red to LightBrown
dark_brown_lower_color = np.array([138,103,10], np.uint8) # from Green to Dark Brown
dark_brown_upper_color = np.array([158,123,90], np.uint8) # from Green to Dark Brown
light_blue_lower_color = np.array([77,93,171], np.uint8) # from Purple to Light Blue
light_blue_upper_color = np.array([97,113, 251], np.uint8) # from Purple to Light Blue
```

```python
# New Lines
lightBrown = cv2.cvtColor(res, light_brown_lower_color, light_brown_upper_color)
darkBrown = cv2.cvtColor(res,)
```

```python
cv2.imwrite("scrapped_colors.jpg", res)
```
